```python
from pymatgen.core.structure import Structure
import numpy as np
```

```python
#TODO combined rotations, coverage
#create (h,k,l) slab(params: number of layers and vacuum space) surface given the bulk structure
def adsorb_mol(d, i, j, angle_rot, Struct_Ligand, Struct_slab, fout_name):
#        d = distance_above_site
#        i = Ligand_bridge_index
#        j = Slab_site_index
        Ligand_site = Struct_Ligand.cart_coords[i]
        Slab_site = Struct_slab.cart_coords[j]
        Adsorption_site = np.array([Slab_site[0], Slab_site[1], Slab_site[2] + d])
        pid = (np.pi)/180
        Ox = pid*angle_rot[0]
        Oy = pid*angle_rot[1]
        Oz = pid*angle_rot[2]
     #   print "Adsorption_site"
      #  print Adsorption_site
        Trans_vector = Adsorption_site - Ligand_site
       # print "Translation vector: "
     #   print Trans_vector
        C_orig = Struct_Ligand.cart_coords
      #  print "Original Coords of Ligand"
      #  print C_orig
        length = len(C_orig)
        x = 0
        y = 0
        R_vector = [[0 for x in xrange(3)] for x in xrange(length)]
      #  print "R_vector initial: "
      #  print R_vector
        Rot_vector = [[0.0 for y in xrange(3)] for y in xrange(length)]
       # print "Rot_vector inital: \n", Rot_vector
        rz_vector = [[0 for m in xrange(3)] for m in xrange(length)]
        ry_vector =[[0 for m in xrange(3)] for m in xrange(length)]
        rx_vector =[[0 for m in xrange(3)] for m in xrange(length)]
       # print "\n", rz_vector, ry_vector, rx_vector
        x = 0
        for x in list(range(length)):
                if x == i:
                    continue
                R_vector[x] = Struct_Ligand.cart_coords[x] - Struct_Ligand.cart_coords[i]
       #         print "\n R_vectors: "
       #         print R_vector[x]
                a = R_vector[x][0]
                b = R_vector[x][1]
                c = R_vector[x][2]
        #        print "\n components of R_vector: "
         #       print a, b, c
                  # z-axis rotation
#                rz_vector[x][0] = (a*np.cos(Oz)) - (b*np.sin(Oz)) - a
#                rz_vector[x][1] = (a*np.sin(Oz)) + (b*np.cos(Oz)) - b
#                rz_vector[x][2] = 0
                # for y - axis rotation
                ry_vector[x][0] = (a*np.cos(Oy)) + (c*np.sin(Oy)) - a
                ry_vector[x][2] = (c*np.cos(Oy)) - (a*np.sin(Oy)) - c
                ry_vector[x][1] = 0
                # for x - axis rotation
#               rx_vector[x][1] = (c*np.cos(Ox)) - (b*np.sin(Ox)) - b
#               rx_vector[x][2] = (c*np.sin(Ox)) + (b*np.cos(Ox)) - c
#               rx_vector[x][0] = 0
     #           print "\n rx, ry and rz: \n"
     #           print rx_vector[x], ry_vector[x], rz_vector[x]
                Rot_vector[x][0] = rx_vector[x][0] + ry_vector[x][0] + rz_vector[x][0]
                Rot_vector[x][1] = rx_vector[x][1] + ry_vector[x][1] + rz_vector[x][1]
                Rot_vector[x][2] = rx_vector[x][2] + ry_vector[x][2] + rz_vector[x][2]
      #          print "\n Rotational Vector: "
      #          print Rot_vector[x]
        New_ligand_coords = [[0.0 for y in xrange(3)] for y in xrange(length)]
        New_ligand_coords[i] = C_orig[i] + Trans_vector
 #       outfile.write(s)
 #       x = 0
 #       print title
        for x in list(range(length)):
#               outfile.write(Struct_Ligand.species)
            if x == i:
                continue
  #              outfile.write("\n")
            New_ligand_coords[x] = C_orig[x] + Trans_vector + Rot_vector[x]
       #     print New_ligand_coords[x]
     #           outfile.write(s)
       # print New_ligand_coords
        Array_New_ligand_coords = np.array(New_ligand_coords)
        Coords_struct_1 = Struct_slab.cart_coords
        Coords_struct_2 = Array_New_ligand_coords
       # print np.size(Coords_struct_1)
        #print np.size(Coords_struct_2)
       # Reshape_new_struct_coords = []
        Reshape_Coords_1= np.reshape(Coords_struct_1, np.size(Coords_struct_1)) 
        Reshape_Coords_2= np.reshape(Coords_struct_2, np.size(Coords_struct_2))
        Reshape_new_struct_coords = np.append(Reshape_Coords_1, Reshape_Coords_2)
        New_struct_length = len(Struct_Ligand.species) + len(Struct_slab.species)
       # type(Reshape_new_struct_coords)
        # New_struct_coords = []
        # np.reshape(Reshape_new_struct_coords, (34, 3))
        New_struct_coords = np.reshape(Reshape_new_struct_coords, (New_struct_length, 3))
        New_struct_species = Struct_slab.species + Struct_Ligand.species
       # New_struct_length = len(Struct_Ligand.species) + len(Struct_slab.species)
        New_struct = Structure(Struct_slab.lattice, New_struct_species, New_struct_coords, coords_are_cartesian=True)
        Sort_struct = New_struct.get_sorted_structure()
        Sort_struct.to("poscar", fout_name)
        print "Successful write out! "
```

```python
distance_above_site = 3.0
Ligand_bridge_index = 1
Slab_site_index = 17
angle_rot = [0.0, 0.0, 0.0]

#Ligand_Poscar)
Struct_Ligand = Structure.from_file("/home/josh/Research/POSCARs/POSCAR_ins/POSCAR_methanethiol_relaxed_E600.vasp") 

#Slab_Poscar
Struct_slab = Structure.from_file("/home/josh/Research/POSCARs/POSCAR_PbS100_1_thiol_3.0A_cart.vasp")  

fout_name = "/home/josh/Research/POSCARs/POSCAR_PbS100_2_thiol_3.0A.vasp"

# call adsorb_mol
adsorb_mol(distance_above_site, Ligand_bridge_index, Slab_site_index, angle_rot, Struct_Ligand, Struct_slab, fout_name)
```
